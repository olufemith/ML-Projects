# Car Accident Severity Prediction Model

## Problem Statement

Some of the major predisposing factors to car accidents are wearther and road conditions. Weather conditions such as high winds and rains can decrease visibility, and make vavigation on roads more tricky and prone to accidents. Also, road conditions and distance of travel are important factors that have safety implications on roads.
To ensure better safety of human lives and properties on road, it is necessary to find ways to inform travellers of the potential risks of a particular trip, based on road and weather conditions at any point in time. This is the objective of the present project, which seeks to develop a model based on historic data on road accidents, that can be used to predict possibilities of road accidents before they happen. Such a tool will help road users to make informed decisions about planned trips to avoid accidents and losses.
The target audience for this project is the local Seattle government, police organization, health system and institutions such as car insurance and road users. The model and its results will provide the government and road users with meaningful data to reduce car accidents and related injuries in the city by enabling the target audience to make meaningful decisions.

## The Data Used

The data to be used for this project is a labelled dataset based on accident severity for Seattle city, shared in the lecture materials for the Applied Data Science Capstone course. The data has 37 attributes (features), and a total of 194,673 observations.
Accidents in the data are categorized into two broad severity types which are:
1. Injury collision and 
2. Property damage only collision
Represented as severity codes 2 and 1 respectively.
The data has been provided by the Seattle Police Department since 2004 and consists of 37 independent variables and 194.673 lines.
To find out which factors impact the level of injuries in accidents, the "SEVERITYCODE" feature is used as the dependent variable, while "WEATHER" which defines the weather at the time of the collision, "ROADCOND", "LIGHTCOND" which describes the road and light conditions and "SPEEDING", which describes the speed of the driver are used as independent variables.
"SEVERITYCODE" contains numbers from 0 to 4 corresponding to different severity levels of accident as follows:
0: Little to no injured.
1: Very Low injured.
2: Low injured.
3: Mild injured.
4: High injured.
